```python
import numpy as np

from full_qubo_solver import FullQuboSolver
from average_partition_solver import AveragePartitionSolver
```

```python
n = 2
m = 2
```

```python
xc = (np.random.rand(n + 1) - 0.5) * 10
yc = (np.random.rand(n + 1) - 0.5) * 10
instance = np.zeros((n + 1, n + 1))
for ii in range(n + 1):
    for jj in range(ii + 1, n + 1):
        instance[ii, jj] = (xc[ii] - xc[jj]) ** 2 + (yc[ii] - yc[jj]) ** 2
        instance[jj, ii] = instance[ii, jj]
```

```python
vrp = FullQuboSolver(n, m, instance)
vrp.build_quadratic_program()
print(vrp.qp.export_as_lp_string())
```

```python
vrp.solve()
```

```python
print(vrp.result)
```

```python
a = vrp.solution.reshape(-1)
b = vrp.variables.reshape(-1)
c = [b[i] for i in range(len(b)) if a[i] == 1]
print(c)
```
